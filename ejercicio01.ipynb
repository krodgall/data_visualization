```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### PREGUNTA 1:

```python
tips['tips_pct(%)'] = ((tips['tip'] / tips['total_bill'])*100)
```

```python
tips.head()
```

```python
grouped_tips_by_size = tips.groupby('size', as_index=False)
result = grouped_tips_by_size.agg({'tips_pct': 'mean'})
result['tips_pct']=result['tips_pct'].round(2)
result
```

### PREGUNTA 2:

```python
grouped_tips_by_time = tips.groupby('time', as_index=False)
result = grouped_tips_by_time.agg({'tips_pct': 'mean'})
result['tips_pct']=result['tips_pct'].round(2)
result
```

```python
grouped_tips_by_day = tips.groupby('day', as_index=False)
result = grouped_tips_by_day.agg({'tips_pct': 'mean'})
result['tips_pct']=result['tips_pct'].round(2)
result
```
