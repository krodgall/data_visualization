```python
import pandas as pd
import matplotlib.pyplot as plt
from plotnine import *
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
(ggplot(titanic, aes(x='Age'))
        + geom_histogram(breaks=range(0,80, 5))
        + theme_minimal())
```

```python
fig, ax = plt.subplots()

x = titanic['Age']

ax.hist(x, bins=range(0, 80, 5), edgecolor='white', zorder=2)

ax.grid(axis='y')
ax.set_title('Distribuci√≥n de edades')
ax.spines[:].set_visible(False) # Quitar los marcos
ax.yaxis.set_tick_params(length=0)

plt.show()
```

```python
import seaborn as sns
```

```python
sns.kdeplot(titanic, x='Age', fill=True, clip=(0,80), bw_adjust=1) #bw_adjust ajusta la suavidad, por defecto es igual a 1 y est√° bien
plt.show()
```

```python
(ggplot(titanic, aes(x='Age'))
    + geom_density(bw=5)
    + theme_minimal()
 )
```

```python
fig, ax = plt.subplots()

x1 = titanic.loc[lambda df: df['Sex'] == 'male', 'Age']
x2 = titanic.loc[lambda df: df['Sex'] == 'female', 'Age']

ax.hist(x1, bins=range(0,80,5), edgecolor='white', alpha=0.5, zorder=2)
ax.hist(x2, bins=range(0,80,5), edgecolor='white', alpha=0.5)

plt.show()
```

```python
sns.kdeplot(titanic, x='Age', hue='Sex', fill=True, clip=(0,80), bw_adjust=1) #bw_adjust ajusta la suavidad, por defecto es igual a 1 y est√° bien
plt.show()
```

```python
(ggplot(titanic, aes(x='Age', fill='Sex'))
    + geom_density(alpha=0.5)
    + theme_minimal()
 )
```

```python
import numpy as np
```

```python
scores = np.random.normal(80, 10, 50)
scores[scores > 100] = 100
```

```python
data = (pd.DataFrame({'scores': scores})
    .sort_values('scores')
    .reset_index(drop=True)
)
```

```python
(ggplot(data, aes(x='scores'))
    + stat_ecdf()
    + theme_minimal()
    )
```

```python
sns.ecdfplot(data=data, x='scores')
plt.show()
```
